```python
x0 = 0.
sigma_x0 = 1.
def gera_dados(M, N):
    x =[]
    sigma_x=[]
    for i in range(M):
        vec = x0 + np.random.randn(N) * sigma_x0
        x.append(vec)
        sigma_vec = np.sqrt(1/(N-1) * np.sum((vec - np.mean(vec))**2))
        sigma_x.append(sigma_vec)
    return np.asarray(x), np.asarray(sigma_x)
N_vec = [100, 10, 5, 4, 3, 2]
for element in N_vec:
    x, sigma_x = gera_dados(10000, element)
    m =[]
    M = []
    for i in range(len(sigma_x)):
        m.append(np.mean(sigma_x[i]))
        M.append(np.median(sigma_x[i]))

    print('N = ', element, np.mean(sigma_x),'com incerteza:', np.sqrt(1/(element - 1) * np.sum((sigma_x - np.mean(sigma_x))**2))/np.sqrt(10000), np.median(sigma_x), len(np.argwhere(sigma_x < sigma_x0)))
    pl.hist(sigma_x,bins=50,label='s')
    pl.legend(fontsize=16)
    pl.show()
```
