```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as pl
from scipy.optimize import curve_fit
from scipy.stats import chi2
import pandas as pd
import cmath
```

```python
pl.rcParams['figure.figsize'] = (16, 8)
pl.rc('xtick', labelsize=16)
pl.rc('ytick', labelsize=16)
```

```python
data = pd.read_csv("polarizador 45 graus.csv", sep=",")
```

```python
x = np.asarray(data['Angle (graus)'])
y = np.asarray(data['Voltage (V)'])
sigma_y = np.asarray(data['Uncertaity (V*5/1023)'])
sigma_x = np.asarray(data['Uncertainty (graus)'])
```

```python
#Modelos
def E(x, A, B, fase):
    x = x * np.pi/180
    return A*np.cos(x+fase)**2 + B
```

```python
fit=curve_fit(f=E, xdata=x, ydata=y, sigma=sigma_y, p0=[2,3, 45], absolute_sigma=True)
ans, cov = fit
print(ans, '\n\n com erros=', np.sqrt(np.diag(cov)))
print('\nCom covariancia=\n', cov)
```

```python
chi2.interval(0.90, len(x)-len(ans)) # recebe alpha e degrees of freedom
```

```python
xplots = np.linspace(0.,180.,500) # vetor para plotar o modelo 
```

```python
np.min(y)
```

```python
pl.errorbar(x, y, xerr = sigma_x, yerr=sigma_y, fmt='.',label=r'Experimental data')
pl.plot(xplots, E(xplots, ans[0], ans[1], ans[2]), label=r'$A\star\cos^2(\theta + fase) + B$')
pl.xlabel(r'Ângulo (º)', fontsize=16)
pl.ylabel(r'Voltagem (V)', fontsize=16)
pl.ylim(0,5)
pl.legend(loc="lower left",fontsize = 16)
pl.axvline(ans[2], c='r', ls=':')
pl.grid(ls=':',c='k')
pl.axhline(np.mean(y), c='r', ls=':')
pl.title('Polarização elíptica', fontsize = 16)
pl.savefig('Polarização Elíptica.png')
```
