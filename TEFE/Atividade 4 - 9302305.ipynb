## ExercÃ­cio 1

```python
N = 50 #50 medidas por experimento
M = 10000 #vezes a realizar o experimento
def Experimento(M):
    resultado = np.zeros(M)
    sigma_a = 28.
    sigma_s = 3.
    for k in range(M):
        erro_s = np.random.randn(1)*sigma_s
        for i in range (N):
            x = 100.0 + np.random.randn(N)*sigma_a + erro_s
            xm = np.mean(x)
        resultado[k] = xm
    return resultado

resultado= Experimento(M)
```

```python
for i in range(len(resultado)):
    print('resultado = %.5f'%(resultado[i]))
```

```python
incerteza = np.std(resultado)
incerteza
```

```python
len(np.argwhere(resultado>100.))
```

```python
len(np.argwhere(abs(resultado-100.)<incerteza))
```
