### Exercício 2

```python
x0 = 500
sigma = 50
N = 100
x=np.zeros((5, 100))
for k in range (5):
    x[k] = x0+np.random.randn(N)*sigma
```

```python
y = np.zeros((5, N))
for k in range (5):
    y[k,0] = x[k,0]
    soma = x[k,0]
    for i in range(1,N):
        soma = soma+x[k,i]
        y[k,i] = soma/(i+1)
```

```python
xplot = np.arange(1,101, 1.)
cores = ['k', 'r', 'b', 'g', 'y']
for k in range(5):
    plt.errorbar(xplot, y[k], c=cores[k], fmt='.',label='Medidas %d' % k)
    plt.plot(xplot, y[k], c=cores[k])
plt.plot(xplot,500*np.ones(100), c='k')
plt.xlabel(r'N.ésima medida X', fontsize=18)
plt.ylabel(r'Média dos N primeiros X', fontsize=18)
plt.legend()
plt.grid(ls=':',c='k')
```
