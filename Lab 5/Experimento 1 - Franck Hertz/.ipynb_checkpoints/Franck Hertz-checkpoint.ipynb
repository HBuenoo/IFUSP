```python
import pandas as pd
import os
import _plotly_future_
from _plotly_future_ import v4_subplots
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
from scipy.optimize import curve_fit
```

```python
medida_numero = os.listdir('/home/hbueno/IFUSP/Lab 5/Experimento 1 - Franck Hertz/Dados') # Aqui vem o caminho dos .csv de todas as medidas tomadas
```

```python
#Modelos
def linear(x, A, B):
    return A*x + B
```

```python
def acha_as_retas(x,y):
    vec_max = []
    vec_min = []
    tensao_max = []
    tensao_min = []
    for i in range(len(y)-1):
        #if i<216:
        if y[i-1] > y[i] < y[i+1]:
            vec_max.append(y[i])
            tensao_max.append(x[i])
            vec_min.append(y[i+1])
            tensao_min.append(x[i+1])
    return np.asarray(vec_max), np.asarray(vec_min), np.asarray(tensao_max), np.asarray(tensao_min)
```

```python
def add_reta(fig, x, y, nome):
    ans, cov =curve_fit(f=linear, xdata=x, ydata=y)
    x_vec = np.linspace(0, 50, 200)
    trace = go.Scatter(
        x=x_vec, y=linear(x_vec, ans[0], ans[1]), mode='lines', name=nome
    )
    fig.append_trace(trace, row=row, col=col)
```

## Plot com um dado

```python
data = pd.read_csv('/home/hbueno/IFUSP/Lab 5/Experimento 1 - Franck Hertz/Dados/03-11_medida07.csv', sep=";")
condicoes = data.iloc[-1].name.replace('Fator de ', '').replace('00', '').replace('26', '') + ' ' + data.iloc[-3].name.replace('00', '')
#data = data.sort_values(data.columns[1])
if (data.iloc[-3].name == 'Canal 3 = 0.4800V'):
    data = data[data.iloc[:, 2] > 0]
    data = data[data.iloc[:, 2] < 10]
else:
    data = data[data.iloc[:, 2] > 0]
    data = data[data.iloc[:, 2] < 15]
```

```python
x = data[data.columns[1]]
y = data[data.columns[2]]
print(data.columns)
y = data['V(calib)'].values
x = data['V(canais)'].values
data.head(1)
```

```python
vec_max,vec_min,tensao_max, tensao_min = acha_as_retas(x, y)
fit=curve_fit(f=linear, xdata=tensao_max, ydata=vec_max)
ans, cov = fit
fit=curve_fit(f=linear, xdata=tensao_min, ydata=vec_min)
ans1, cov2 = fit
```

```python
xplot = np.linspace(0, 50, 200)
trace1 = go.Scatter(
    x=x, y=y, mode='markers', name='Pontos experimentais', marker=dict(size=8, opacity=0.8)
)
trace2 = go.Bar(x=x, y=y)

trace1 = go.Scatter(
    x=x, y=y, mode='lines', name='Pontos experimentais'#, marker=dict(size=8, opacity=0.8)
)
trace2 = go.Scatter(
    x=xplot, y=linear(xplot, ans[0], ans[1]), mode='lines'
)
trace3 = go.Scatter(
    x=xplot, y=linear(xplot, ans1[0], ans1[1]), mode='lines'
)
trace4 = go.Scatter(
    x=tensao_max, y=vec_max, mode='markers', marker=dict(size=8, opacity=0.8), name='vec_max'
)
trace5 = go.Scatter(
    x=tensao_min, y=vec_min, mode='markers', marker=dict(size=8, opacity=0.8), name='vec_min'
)

layout = go.Layout(
    yaxis=dict(title='I'), xaxis=dict(title='V'), showlegend = True
)
fig = go.Figure(data=[trace1, trace2, trace3, trace4, trace5], layout=layout)
fig.show()
```

## Plot com todos

```python
from plotly.subplots import make_subplots
fig=make_subplots(rows=5, cols=2, shared_xaxes=True,shared_yaxes=True, x_title="Tensão [V]", y_title="N(I)")
row=1
col=1
for name in medida_numero:
    #print(name)
    data = pd.read_csv('/home/hbueno/IFUSP/Lab 5/Experimento 1 - Franck Hertz/Dados/' + name, sep=';')
    data = data.sort_values(data.columns[0])
    if (data.iloc[-3].name == 'Canal 3 = 0.4800V'):
        data = data[data.iloc[:, 2] > 0]
        data = data[data.iloc[:, 2] < 10]
    else:
        data = data[data.iloc[:, 2] > 0]
        data = data[data.iloc[:, 2] < 15]
    x = data['V(canais)'].values
    y = data['V(calib)'].values
    vec_max,vec_min,tensao_max, tensao_min = acha_as_retas(x, y)
    add_reta(fig, tensao_max, vec_max, 'reta dos picos')
    add_reta(fig, tensao_min, vec_min, 'reta dos minimos')
    trace1 = go.Scatter(
        x=x, y=y, name='dados da telinha', mode='markers', marker=dict(size=3, opacity=0.8)
    )
    trace2 = go.Scatter(
        x=tensao_max, y=vec_max, mode='markers', marker=dict(size=8, opacity=0.8), name='cristas'
    )
    trace3 = go.Scatter(
        x=tensao_min, y=vec_min, mode='markers', marker=dict(size=8, opacity=0.8), name='vales'
    )
    fig.append_trace(trace1, row=row, col=col)
    fig.append_trace(trace2, row=row, col=col)
    fig.append_trace(trace3, row=row, col=col)
    row += 1
    if(row>5):
        row = 1
        col = 2

fig.show(filename='Experimento de Franck Hertz',title="Experimento de Franck Hertz")
```
