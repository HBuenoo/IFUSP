## Exercício 3

```python
N=10000
def n_2D(N):
    x = np.random.random(N)
    y = np.random.random(N)
    n = len(np.argwhere(x**2 + y**2 <= 1))
    return n
def n_3D(N):
    x = np.random.random(N)
    y = np.random.random(N)
    z = np.random.random(N)
    n = len(np.argwhere(x**2 + y**2 + z**2<= 1))
    return n
n2d, n3d = [], []
for i in range(1000):
    n2d.append(n_2D(N))
    n3d.append(n_3D(N))
    
A = 4/N * np.asarray(n2d)
B = 6/N * np.asarray(n3d)
```

```python
sigma_n_2d = np.std(n2d)
sigma_A = np.std(A)
sigma_B = np.std(B)
sigma_n_3d = np.std(n3d)
```

```python
print(sigma_n_2d, sigma_n_3d)
```
