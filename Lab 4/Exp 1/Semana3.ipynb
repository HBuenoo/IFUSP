```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as pl
from scipy.optimize import curve_fit
from scipy.stats.distributions import  t
from scipy.stats import chisquare
import scipy
import pandas as pd

pl.rcParams['figure.figsize'] = (16, 5)
pl.rc('xtick', labelsize=16)
pl.rc('ytick', labelsize=16)
```

```python
data = pd.read_csv("/home/hbueno/IFUSP/Lab 4/Exp 1/Semana3.csv", sep=",")
tipo = np.asarray(data.tipo)
posicao = np.asarray(data.posicao)
sigma = np.asarray(data.incerteza)
n = np.asarray(data.n)
```

```python
data
```

```python
#Modelos
def single_slit(x, a):
    wavelenght=6.33*10**-5 # cm
    L = 26.21 #cm
    d = 0.001 #cm
    theta = np.arctan(x/L)
    return a*np.sin(theta)**2 * (wavelenght/(np.pi*d*np.sin(theta)))**2

def linear_fit(x, a, b):
    return  a*x + b

def chi2(obs, esp, erro):
    return np.sum((obs-esp)**2/erro**2)
def chi2red(obs, esp, erro):
    return np.sum((obs-esp)**2/erro**2)/(len(obs)-len(ans))
def residuo(obs, esp):
    return obs - esp
```

```python
single = np.argwhere(tipo==1)[:,0]
double = np.argwhere(tipo==2)[:,0]
rede = np.argwhere(tipo==3)[:,0]
```

```python
fit=curve_fit(linear_fit, posicao[single],n[single], sigma=sigma[single], absolute_sigma=True)
ans, cov = fit
print(ans, '\n\n', np.sqrt(np.diag(cov)))
```

```python
x_plots = np.linspace(0,5,100) # vetor para plotar o modelo 

pl.errorbar(posicao[single], n[single], xerr=sigma[single], fmt='.',c='k',label=r'Experimental data')
pl.plot(x_plots,linear_fit(x_plots, ans[0], ans[1]),c='b',lw=2,label=r'$y=a*x + b$ fit')
pl.xlabel(r'Posição do pico (cm)', fontsize=18)
pl.ylabel(r'N', fontsize=18)
pl.legend(loc="lower right",fontsize = 16)
pl.grid(ls=':',c='k')
pl.savefig('Semana 3 - single_slit.png')
```

```python
pl.errorbar(posicao[single], residuo(n[single], linear_fit(posicao[single], ans[0],ans[1])), xerr = sigma[single], fmt='.', c='k')
pl.axhline(y=0,c='r')
pl.xlabel(r'Posição do pico (cm)', fontsize=18)
pl.ylabel(r'$\sigma^2$ (cm)', fontsize=18)
pl.grid()
```

```python
chi2red(posicao[single], linear_fit(posicao[single], ans[0],ans[1]), sigma[single])
```

```python
residuo(posicao[single], linear_fit(posicao[single], ans[0],ans[1]))
```

```python
crit = scipy.stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = len(posicao[single])-len(ans))   # Df = number of variable categories - 1
crit
```
