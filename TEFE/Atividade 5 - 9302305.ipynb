```python
import numpy as np
import matplotlib.pyplot as pl
import scipy.stats as st
from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
import pandas as pd
```

## Exercício 1

```python
 def dens_prob(x):
        G=1
        A = 1
        return A*(1-abs(x)**G)
    
def gera_x_por_exclusao(n):
    x=np.zeros(n)
    xi = -1
    xf = 1
    A = 1
    nOk = 0
    while (nOk < n):
        xc = xi + (xf - xi)*np.random.rand()
        yc = A*np.random.rand()
        if (yc < dens_prob(xc)):
            x[nOk] = xc
            nOk = nOk + 1
    return x
```

```python
N=10000
x = gera_x_por_exclusao(N)
xm =  np.mean(x)
sx = np.std(x)
sxm = np.std(x)/np.sqrt(N)
freq=0
for i in range(N):
    if (xm-sx<= x[i] <= xm+sx):
        freq = freq + 1
        

print(' x médio:', np.mean(x), '\n desvio padrão de x:', np.std(x), '\n desvio padrão da média:', np.std(x)/np.sqrt(N), '\n freq:', freq, '\n freq relativa:', freq/N)

x_plot = np.linspace(-1,1,N)
pl.plot(x_plot, dens_prob(x_plot), label='density probability function')
pl.hist(x, alpha = 0.5, density=True, label='Generated x')
pl.xlabel(r'x', fontsize=18)
pl.ylabel(r'f(x)', fontsize=18)
pl.legend(loc="upper right",fontsize = 16)
pl.grid(ls=':',c='k')
```
