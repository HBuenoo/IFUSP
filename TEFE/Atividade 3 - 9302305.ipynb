```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
import pandas as pd
```

### Exercício 1

```python
N = 10000
r = 15. + 1.*np.random.randn(N)
x = np.pi*r**2
print('A incerteza de x= %.5f'%np.std(x))
xm = np.mean(x)
sigma_xm = np.std(x)/np.sqrt(N)
print('x médio = %.5f com incerteza = %.5f'%(xm, sigma_xm))
```

### Exercício 2

```python
a = 85. + 10.*np.random.randn(N) # em graus
a = a * np.pi/180. # agora em rad
y = np.sin(a)
print('A incerteza de y=%.5f'%np.std(y))
ym = np.mean(y)
sigma_ym = np.std(y)/np.sqrt(N)
print('y médio = %.5f com incerteza = %.5f'%(ym, sigma_ym))
```

```python
a_plot = np.linspace(45.,135., 50)*np.pi/180
plt.errorbar(a, y,xerr=10.*np.pi/180, yerr = np.std(y), c='k', fmt='.', alpha=0.15)
plt.plot(a_plot, np.sin(a_plot), c='b', alpha =1.0, label='y = sin (a)')
plt.plot(a_plot, np.sin(a_plot)-np.std(y), c='r', alpha =1.0, label='1 sigma abaixo')
plt.plot(a_plot, np.sin(a_plot)+np.std(y), c='g', alpha =1.0,label='1 sigma acima')
plt.xlabel('a (rad)', size=16)
plt.ylabel('sin (a)', size=16)
plt.legend(fontsize=15)
plt.show()
```
